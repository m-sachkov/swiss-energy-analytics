### National electricity consumption

National electricity consumption, including consumption by storage pumps, is an important indicator of the electricity supply. The trend is initially presented based on data reported to Swissgrid by distribution system operators. This data may be incomplete or incorrect and is subsequently corrected. Final consumption indicates how much electricity is consumed by households, commerce, agriculture, services, industry, and transportation. When it comes to the Swiss public’s energy-saving efforts, it is the key metric. The trend is initially displayed based on data reported to Swissgrid by distribution network operators. This data may be incomplete or incorrect and is subject to correction.\nIn this specific dataset, updated values are appended to the existing data rather than overwriting older entries. The most recent entry for each date corresponds to the most current and valid value. The purpose of this structure is to provide a transparent history of data changes.

```python
response = requests.get("https://energiedashboard.ch/api/v1/datasets/stromverbrauch-swissgrid-lv-und-endv/data?offset=0")

data = response.json()
df = pd.DataFrame(data['data'])
```

### Meteo

```python
response = requests.get("https://data.geo.admin.ch/api/stac/v1/collections/ch.meteoschweiz.ogd-smn/items")
data = response.json()

df = pd.DataFrame(data['features'])
df.head()
```

```python
assets = [item['assets'] for item in data['features']]
[(filename, metadata['href']) for asset in assets for (filename, metadata) in asset.items()]
```

```python
assets = [item['assets'] for item in data['features']]
datasets_urls = [
    metadata['href'] 
    for asset in assets
    for metadata in asset.values()
]
datasets_urls
```
